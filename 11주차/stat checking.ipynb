```python
x = [1,3,5,7,9]; tp = np.eye(5); lp = np.eye(5)
fig, ax = plt.subplots(nrows= 5, ncols=5)
for i in range(len(x)):
    for j in range(len(x)):
        x1 = data.loc[data.label == x[i]].APW
        x2 = data.loc[data.label == x[j]].APW
        sns.distplot(x1, ax = ax[i][j])
        sns.distplot(x2, ax = ax[i][j])
        tp[i][j] = st.ttest_ind(x1, x2, equal_var = False)[1]
        lp[i][j] = st.levene(x1, x2)[1]
fig.savefig('stats.jpg')
```
